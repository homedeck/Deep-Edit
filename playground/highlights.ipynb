# Highlight Regularization

```python
from imageio import imread
from matplotlib import pyplot as plt

image = imread("../media/1.jpg")

plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
from numpy import where

mask = where(image > 252., 1., 0.)

plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.imshow(mask)
plt.xticks([])
plt.yticks([])
plt.show()
```
